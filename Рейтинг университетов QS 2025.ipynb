### __Рейтинг университетов QS 2025__

__О наборе данных__

В этом наборе данных представлен рейтинг университетов мира QS 2025 года.

Он охватывает основные показатели и индикаторы, используемые для оценки академического совершенства, международного разнообразия, исследовательского воздействия и трудоустройства университетов по всему миру.

Каждая запись включает в себя рейтинг учреждения, название, местоположение, категорию размера, академические баллы и баллы репутации среди работодателей, соотношение профессорско-преподавательского состава и студентов, цитирование на факультет, долю иностранных преподавателей и студентов, присутствие в международных исследовательских сетях, результаты трудоустройства, показатели устойчивого развития и общий балл QS.

В совокупности эти показатели дают представление о глобальной ситуации в сфере высшего образования, облегчая сравнительный анализ и принятие обоснованных решений для различных заинтересованных сторон, включая студентов, преподавателей, политиков и работодателей.

 0 2025 Rank : Рейтинг QS 2025                       
 1 2024 Rank : Рейтинг QS 2024 г.                      
 2 Institution Name : Название университета               
 3 Location  : Код страны, в которой находится учреждение                      
 4 Location Full : Полное название страны, где находится учреждение         
 5 Size  : Размерная категория учреждения    
 6 Academic Reputation: Оценка, отражающая академический статус учебного заведения.

 8 Faculty Student: Соотношение преподавательского состава и студентов, служащее показателем качества преподавания        
 9 Citations per Faculty : 
Показатель влияния и качества исследования, указывающий количество цитирований, полученных на одного преподавателя.         
 10 International Faculty : Процент иностранных преподавателей, что отражает разнообразие академического состава         
 11 International Students : Процент иностранных студентов, зачисленных в учебное заведение     
 12 International Research Network : Оценка, отражающая степень международного исследовательского сот
д ничества
 13  Employment Outcomes  : Оценка, указывающая на трудоустройство и карьерные перспективы выпускнико.        
 14 Sustainability : Оценка, устойчивого развити.  
 15 QS Overall Score  Общий балл QS, присваиваемый университету на основе взвешенной комбинации показателей.

```python
#извлекаем только начальное значение ранга, игнорируя диапазон.
def split_rank(X):
    if type(X) == str:
        if '-' in X:
            X = X.split('-')[0]
        if '+' in X:    
            X = X.split('+')[0]
    return X
```

```python
df['2025 Rank'] = df['2025 Rank'].apply(split_rank)
df['2025 Rank'] = df['2025 Rank'].astype(int)
df['2024 Rank'] = df['2024 Rank'].apply(split_rank)
df['2024 Rank'] = df['2024 Rank'].astype(int)
df.info()
```

__Модель линейной регрессии__

```python
df2 = df.drop(['Institution Name','Location', 'QS Overall Score'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2['Location Full'] = le.fit_transform(df2['Location Full'])
df2['Size'] = le.fit_transform(df2['Size'])
```

```python
df2.info()
```

```python
X = df2.drop(['2025 Rank'],axis=1)
y = df2['2025 Rank']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X),columns=X.columns)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#обучаем модель линейной регрессии на обучающих данных 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
#вывод значения R²
print('R2 : ',lr.score(X_test,y_test))
```

__Вывод.__ Значение R² ближе к 1 указывает на то, что модель хорошо объясняет изменчивость данных.

```python
# Преобразуем столбец «Общий балл QS» в числовой формат, приведя ошибки к NaN.
df['QS Overall Score'] = pd.to_numeric(df['QS Overall Score'], errors='coerce')

# Создадим двоичную целевую переменную для классификации
# Предположим, мы хотим разделить университеты на две категории на основе их общего балла QS:
# 1. Лучшие университеты (общий балл QS >= 90)
# 2. Другие университеты (общий балл QS <90)

df['Top University'] = (df['QS Overall Score'] >= 90).astype(int)

# Отобразим первые несколько строк, чтобы проверить новый столбец.
df[['Institution Name', 'QS Overall Score', 'Top University']].head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Выбираем числовые столбцы для кластеризации
numerical_cols = ['Academic Reputation', 'Employer Reputation', 'Faculty Student', 'Citations per Faculty', 
                  'International Faculty', 'International Students', 'International Research Network', 
                  'Employment Outcomes', 'Sustainability']

# Извлекаем числовые данные для кластеризации
X_clustering = df[numerical_cols]

# Применяем кластеризацию KMeans
kmeans = KMeans(n_clusters=6, random_state=42)
clusters = kmeans.fit_predict(X_clustering)

# Добавляем метки кластера в фрейм данных
df['Cluster'] = clusters

# Строим кластеры
plt.figure(figsize=(10, 6))
plt.scatter(df['Academic Reputation'], df['Employer Reputation'], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.title('Clustering of Universities based on Academic and Employer Reputation')
plt.xlabel('Academic Reputation')
plt.ylabel('Employer Reputation')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()


# Отображаем первые несколько строк фрейма данных с метками кластера.
df.head()
```
